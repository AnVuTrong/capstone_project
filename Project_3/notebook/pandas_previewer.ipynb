```python
# Rename the columns
df_restaurants.rename(columns={'ID': 'IDRestaurant'}, inplace=True)
df_restaurants.rename(columns={'Time': 'OpeningTime'}, inplace=True)
df_reviews.rename(columns={'Time': 'ReviewTime'}, inplace=True)
```

```python
# Check for duplicates that has the same Restaurant, Address but different IDRestaurant.
duplicates = df_restaurants[df_restaurants.duplicated(subset=['Restaurant', 'Address'], keep=False)]
duplicates
```

```python
# Create a mapping of duplicate IDs to the first occurrence
id_map = {}
for _, group in duplicates.groupby(['Restaurant', 'Address']):
    first_id = group['IDRestaurant'].iloc[0]
    for id_ in group['IDRestaurant']:
        id_map[id_] = first_id
```

```python
id_map
```

```python
# Update df_reviews to replace old IDRestaurant values with new ones
df_reviews['IDRestaurant'] = df_reviews['IDRestaurant'].replace(id_map)
```

```python
# Remove duplicates from df_restaurants, keeping the first occurrence
df_restaurants = df_restaurants.drop_duplicates(subset=['Restaurant', 'Address'], keep='first')
df_restaurants
```

```python
# Group by IDRestaurant and count the number of reviews
review_counts = df_reviews.groupby('IDRestaurant').size().reset_index(name='ReviewCount')
review_counts
```

```python
# Merge the counts with the df_restaurants DataFrame
df_restaurants_reviews = df_restaurants.merge(review_counts, on='IDRestaurant', how='left')
df_restaurants_reviews
```

```python
# Merge the two dataframes
df = pd.merge(df_reviews, df_restaurants_reviews, left_on='IDRestaurant', right_on='IDRestaurant')
```

```python
# ReviewTime format is dd/mm/yyyy hh:mm, convert it to datetime.
df['ReviewTime'] = pd.to_datetime(df['ReviewTime'], format='%d/%m/%Y %H:%M')

# OpeningTime format is hh:mm - hh:mm, separate it into OpeningTime and ClosingTime.
df['ClosingTime'] = df['OpeningTime'].str.split(' - ').str[1]
df['OpeningTime'] = df['OpeningTime'].str.split(' - ').str[0]
```

```python
# Price range format is xx.xxx - xx.xxx, separate it into MinPrice and MaxPrice.
df['MaxPrice'] = df['Price'].str.split(' - ').str[1]
df['MinPrice'] = df['Price'].str.split(' - ').str[0]

# Remove the Price column
df.drop('Price', axis=1, inplace=True)
```
